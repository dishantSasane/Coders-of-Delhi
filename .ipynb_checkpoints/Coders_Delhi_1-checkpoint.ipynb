```python
import json
data = {
    "users": [
        {"id": 1, "name": "Amit", "friends": [2, 3], "liked_pages": [101]},
        {"id": 2, "name": "Priya", "friends": [1, 4], "liked_pages": [102]},
        {"id": 3, "name": "Rahul", "friends": [1], "liked_pages": [101, 103]},
        {"id": 4, "name": "Sara", "friends": [2], "liked_pages": [104]}
    ],
    "pages": [
        {"id": 101, "name": "Python Developers"},
        {"id": 102, "name": "Data Science Enthusiasts"},
        {"id": 103, "name": "AI & ML Community"},
        {"id": 104, "name": "Web Dev Hub"}
    ]
}

with open("data.json","w") as file:
    json.dump(data, file)
```

```python
file.close()
```

```python
def load_data(filename):#function define to load data
    ''' This funcion is used to load data from file'''
    with open(filename, "r") as file: # use with open to automatically close when done
        data = json.load(file)          #json.load is used to load data in python
    return data
```

```python
data = load_data("data1.json")
print (data)
```

```python
import json 

data1 =  {
    "users": [
        {"id": 1, "name": "Amit", "friends": [2, 3], "liked_pages": [101]},
        {"id": 2, "name": "Priya", "friends": [1, 4], "liked_pages": [102]},
        {"id": 3, "name": "Rahul", "friends": [1], "liked_pages": [101, 103]},
        {"id": 4, "name": "Sara", "friends": [2], "liked_pages": [104]}
    ],
    "pages": [
        {"id": 101, "name": "Python Developers"},
        {"id": 102, "name": "Data Science Enthusiasts"},
        {"id": 103, "name": "AI & ML Community"},
        {"id": 104, "name": "Web Dev Hub"}
    ]
}

with open("data1.json","w") as file:
    json.dump(data1, file, indent = 4)
```

```python
#function to display info from data
def display_users(data):   # Name{name} ; ID - {id} ; Friends - {friends} ; Liked Pages - {liked pages}
    print ("Users and their connections \n")
    #loop to print user information
    print("User Info")
    for users in data ['users']:
        print(f"ID- {users['id']}; Name: {users['name']}; Friends- {users['friends']}; Liked Pages -{users['liked_pages']}")
    print ("\n")

    #Loop to print pages information
    print ("Pages Info")
    for pages in data['pages']:
        print(f"Page Id - {pages['id']} is for {pages['name']} page.")

display_users(data1)
```
