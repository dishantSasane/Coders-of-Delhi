```python
#Load the data
import json
```

```python
def clean_data(filename):
    data['users'] = [users for users in data['users'] if users['name'].strip() ]
    
    for users in data['users']:
        users['friends'] = list(set(users['friends']))

    data['users'] = [users for users in data['users'] if users['friends'] and users['liked_pages']]

    seen_id = set()
    unique_page = []
    for pages in data['pages']:
        if pages['id'] not in seen_id:
            unique_page.append(pages)
            seen_id.add(pages['id'])
# 
    data['pages'] = unique_page

    return data



with open("data1.json" , "r") as file:
    data = json.load(file)

cleaned_data = clean_data(data)

with open("cleaned_data.json", "w") as file:
    json.dump(cleaned_data, file, indent=4)
    
print("Data has been cleaned")
```
