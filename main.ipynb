```python
import json
```

```python

def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f) 
    return data
```

```python
data = load_data("data.json")
data
```

```python
# Now we will print the data in an organised manner

def display_users(data):
    for users in data['users']:
        print(f"ID-{users["id"]}, Name- {users["name"]}, Friends-{users["friends"]}, Liked Pages-{users["liked_pages"]}")
        


def display_pages(data):
    for pages in data["pages"]:
        print(f"ID:- {pages["id"]}, Name:- {pages["name"]}")
```

```python
display_users(data)

```

```python
display_pages(data)
```

```python
# Display connections and their liked pages

def display_data(data):
    for users in data["users"]:
        print(f"{users["id"]}: {users["name"]} is friends with {users["friends"]} and likes the pages ({users["liked_pages"]}) ")
    
    print ("\nPages Information")

    for pages in data["pages"]:
        print(f"{pages["id"]} - {pages["name"]}")

```

```python

display_data(data)
```
